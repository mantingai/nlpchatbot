## <font color= grey> Deal with Imbalance Data<a class="anchor" id="4.0"></a>

In the realm of medical text classification, where classes may denote distinct specialties or conditions, maintaining a balanced dataset is crucial to mitigate the risk of model overfitting to the predominant class. 

Overfitting arises when the model becomes excessively attuned to the prevalent class, leading to suboptimal performance when handling less prevalent classes. Conversely, underfitting arises when the model fails to capture the complexities of the data, leading to inadequate performance across all classes. 

A balanced dataset facilitates a more robust training process, helping the model generalize well to various classes without sacrificing performance on specific subsets.

>**Remove values that has less than 250**<a class="anchor" id="4.1"></a>

We plan to exclude classes with insufficient data during the training phase because a lack of textual data for certain classes could potentially result in underfitting. Underfitting occurs when a model fails to grasp the intricacies of a particular class due to a scarcity of representative examples. 

By removing such classes with inadequate data, we aim to enhance the model's ability to effectively learn and generalize across all classes, preventing the detrimental effects of underfitting on specific subsets.

>**Data Augmentation**<a class="anchor" id="4.2"></a>

**Data Augmentation** is a technique commonly used in computer vision.
In image dataset, It involves creating new images by transforming(rotate, translate, scale, add some noise) the ones in the data set. 
<br><br>
When applied to text data, data augmentation can be achieved by tokenizing a document into sentences, shuffling them, and then rejoining to produce new texts. Another approach involves substituting words, such as adjectives or verbs, with their synonyms to generate varied text while preserving the same meaning. Utilizing pre-trained word embeddings or resources like NLTK‚Äôs WordNet can assist in identifying synonyms.
<br><br>
The primary objective of data augmentation in text is twofold: 

1. It allows for the expansion of the dataset, enabling the training of models with an adequate amount of data. 

2. It helps balance out the distribution of classes, particularly addressing minority classes, thereby enhancing the model's ability to generalize across diverse textual patterns.

```python
# load nltk's English stopwords as variable called 'stop' and don't find synonym of those words.
stop = nltk.corpus.stopwords.words('english')
```

```python
# Tokenizing sentence into token for finding synonym.
def make_tokenizer(texts):
    from keras.preprocessing.text import Tokenizer
    t = Tokenizer()
    t.fit_on_texts(texts)
    return t

tokenizer = make_tokenizer(df['transcription']) 

X = tokenizer.texts_to_sequences(df['transcription'])

from keras_preprocessing.sequence import pad_sequences
X = pad_sequences(X, 70)
```

```python
# Dictionary of word index
index_word = {}
for word in tokenizer.word_index.keys():
    index_word[tokenizer.word_index[word]] = word
```

```python
# word list
words = [value for key, value in index_word.items()]
```

```python
# Function to find synonym of words 
import spacy
nlp = spacy.load('en_core_web_sm')
def check_lemma(t,w) :
    r = [d for d in t if (nlp(d.text)[0].lemma_ != nlp(w.text)[0].lemma_)]
    return r

def get_word_synonym(word):
    filtered_words = [w for w in word.vocab if (not w.lower_ in stop) and w.is_lower == word.is_lower and w.prob >= -15] ## (not w.lower_ in stop) and
    similarity = sorted(filtered_words, key=lambda w: word.similarity(w), reverse=True)
    filtered_similarity = check_lemma(similarity[:30], word)
    return filtered_similarity[:3]
```

```python
# Synonym dictionary
synonym_dict = {}

def check_oos(word):
    return word.lower() not in nlp.vocab

for word in words:
    if (not check_oos(word)) :
        synonym_dict.update({word : tuple([w.lower_ for w in get_word_synonym(nlp.vocab[word])])})
        #print(word, " : ", [w.lower_ for w in get_word_synonym(nlp.vocab[word])])
```

```python
# Only consider filtered synonym
import collections
value_occurrences = collections.Counter(synonym_dict.values())

filtered_synonym = {key: value for key, value in synonym_dict.items() if value_occurrences[value] == 1}
```

```python
# Function for augmenting data by replacing words with synonym using spaCy
import re
import random
sr = random.SystemRandom()
split_pattern = re.compile(r'\s+')
def data_augmentation(message, aug_range=1) :
    augmented_messages = []
    for j in range(0,aug_range) :
        new_message = ""
        for i in filter(None, split_pattern.split(message)) :
            new_message = new_message + " " + sr.choice(filtered_synonym.get(i,[i]))
        augmented_messages.append(new_message)
    return augmented_messages
```

```python
# Dictionary for intent count
intent_count = df.medical_specialty.value_counts().to_dict()
```

```python
# Get max intent count to match other minority classes through data augmentation
import operator
max_intent_count = max(intent_count.items(), key=operator.itemgetter(1))[1]
```

```python
import numpy as np
import math
import tqdm

newdf = []  # Initialize as a list

for intent, count in intent_count.items():
    count_diff = max_intent_count - count
    multiplication_count = math.ceil((count_diff) / count)
    if multiplication_count:
        old_message_df = pd.DataFrame(columns=['transcription', 'medical_specialty'])  # Initialize as an empty DataFrame with columns
        new_message_df = pd.DataFrame()
        for message in tqdm.tqdm(df[df["medical_specialty"] == intent]["transcription"]):
            dummy1 = pd.DataFrame([message], columns=['transcription'])
            dummy1["medical_specialty"] = intent
            old_message_df = pd.concat([old_message_df, dummy1])  # Concatenate the DataFrames

            new_messages = data_augmentation(message, multiplication_count)  # Pass multiplication_count as an argument
            dummy2 = pd.DataFrame(new_messages, columns=['transcription'])
            dummy2["medical_specialty"] = intent
            new_message_df = pd.concat([new_message_df, dummy2])  # Concatenate the DataFrames

        new_message_df = new_message_df.take(np.random.permutation(len(new_message_df))[:count_diff])

        newdf.append(old_message_df)  # Append DataFrames to the list
        newdf.append(new_message_df)  # Append DataFrames to the list
    else:
        newdf.append(df[df["medical_specialty"] == intent])  # Append DataFrames to the list

df = pd.concat(newdf)  # Convert the list of DataFrames to a single DataFrame
```

## <font color= grey> Text Normalisation<a class="anchor" id="5.0"></a>

>**Lower Case**<a class="anchor" id="5.1"></a>

```python
# # To convert transcription into lowercase
def lower(df, attribute):
    df.loc[:,attribute] = df[attribute].apply(lambda x : str.lower(x))
    return df
df = lower(df,'transcription')
df.head()
```

>**Remove Punctuations**<a class="anchor" id="5.2"></a>

```python
def remove_punc_keep_digits(df, attribute):
    df[attribute] = df[attribute].apply(lambda x: re.sub(r'[^\w\d.]', ' ', x))
    df[attribute] = df[attribute].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
    return df

df = remove_punc_keep_digits(df, 'transcription')
df.head()
```

>**Removing Stop Words**<a class="anchor" id="5.3"></a>

```python
# # Define the function to remove stop words
def remove_stop_words(text):
    stop = stopwords.words('english')
    return ' '.join([word for word in text.split() if word.lower() not in stop])

# # Apply the function to the 'transcription' column
df['cleaned_transcription'] = df['transcription'].apply(remove_stop_words)
df.head()
```

>**Tokenisation**<a class="anchor" id="5.4"></a>

```python
# to tokenise transcription

tk =WhitespaceTokenizer()

def tokenise(df, attribute):
    df['cleaned_transcription'] = df.apply(lambda row: tk.tokenize(str(row[attribute])), axis=1)
    return df

df.head()
```

>**Lemmatize<a class="anchor" id="5.5"></a>**

```python
nlp = spacy.load("en_core_web_sm")

def lemmatize_column(column, nlp):
    lemmatized_columns = []

    for text in column:
        doc = nlp(text)
        lemmatized_text = " ".join([token.lemma_ for token in doc])
        lemmatized_columns.append(lemmatized_text)

    return lemmatized_columns

lemmatize_column(df['cleaned_transcription'], nlp)

df.head()
```

>**All-in-One function**<a class="anchor" id="5.6"></a>

This function is intended for subsequent vectorization as analyzer.

```python
nlp = spacy.load("en_core_web_sm")

def clean_text(text):
    # Convert text to lowercase
    text = str.lower(text)
    
    # Remove standalone dots
    text = re.sub(r'\b\.(?!\d)', ' ', text)
    
    # Tokenize the text into words using spaCy
    doc = nlp(text)
    tokens = [token.text for token in doc if not token.is_stop and token.is_alpha]
    
    # Join the tokens into a single string
#     cleaned_text = ' '.join(tokens)
    
    return tokens

def lemmatize_tokens(tokens, nlp):
    doc = nlp(" ".join(tokens))
    lemmatized_tokens = [token.lemma_ for token in doc]
    return lemmatized_tokens

# print(clean_text("1.5 medical patient"))
```

## <font color= grey> Test Representation<a class="anchor" id="7.0"></a>

For text representation, there are 2 different methods, which includes:
<br><br>
1. **Bag-of-N-grams:**<br><br>

The Bag-of-N-grams method is a text representation technique that focuses on capturing the frequency of contiguous sequences of N words (N-grams) in a document.
<br><br>
How it Works: <br>The document is tokenized into N-grams, and the count or presence of each N-gram is used to create a numerical representation of the text. The order of the words is disregarded, emphasizing the frequency of individual N-grams.<br><br>

Advantages:<br>
1. Simplicity: Bag-of-N-grams is straightforward and easy to implement.
2. Captures Local Context: It captures local context and word patterns within a specified window.<br><br>

Limitations:<br>
1. Loss of Sequence Information: The method loses the sequential order of words, leading to a potential loss of information about word relationships and meaning.
2. Increased Dimensionality: As N increases, the dimensionality of the representation grows, which can be computationally expensive.

<br><br>
2. **TF-IDF (Term Frequency-Inverse Document Frequency):**<br><br>
TF-IDF is a statistical measure used to evaluate the importance of a word in a document relative to a collection of documents (corpus).
<br><br>
How it Works:<br> TF-IDF considers both the frequency of a term within a document (Term Frequency) and its rarity across the entire corpus (Inverse Document Frequency). It assigns higher weights to terms that are frequent within a document but rare in the overall corpus.<br><br>

Advantages:<br>
1. Emphasizes Uniqueness: It emphasizes terms that are distinctive to a document, potentially capturing the document's topical relevance.
2. Mitigates Common Words: Common terms across documents are assigned lower weights, helping to reduce the impact of frequently occurring words.<br><br>

Limitations:<br>
1. Disregards Word Order: Similar to Bag-of-N-grams, TF-IDF does not consider the order of words, potentially overlooking certain linguistic nuances.
2. Limited Semantic Understanding: It may struggle to capture semantic relationships between words.

In order to enhance the balance within each set, we opted to employ **SMOTE** as a means of further refining our dataset.

### **What is Smote?**
SMOTE, which stands for Synthetic Minority Over-sampling Technique, is a resampling technique used in machine learning to address the class imbalance problem. The main goal of SMOTE is to balance class distribution by creating synthetic examples of the minority class, thereby preventing the model from being biased toward the majority class.

### **How does it work?**
SMOTE works by generating synthetic examples for the minority class by interpolating between existing instances. It creates synthetic samples by taking random combinations of existing minority class instances and generating new instances along the lines connecting them.

### **Why Smote?**
SMOTE is applied after vectorization, specifically to the training set, to address the class imbalance in the transformed feature space.Balancing the classes after vectorization ensures that the synthetic examples generated by SMOTE are represented appropriately in the numerical feature space.

## <font color= grey>Model Evaluation Metrics<a class="anchor" id="2.0"></a>

I have chosen to use F1 score as my main evaluation metric for this NLP Project. Let me explain why:<br><br>

1. **Precision:** <br>In a medical context, precision is essential when the cost of false positives is high. For example, if a model predicts a disease, we want to be confident that the prediction is accurate to avoid unnecessary treatments or interventions.<br><br>

2. **Recall:**<br>In medical scenarios, recall is crucial when the cost of false negatives is high. Missing a positive case (false negative) might mean failing to diagnose a medical condition, leading to delayed treatment and potentially serious consequences.<br><br>

3. **F1 Score:**<br>The F1 score combines precision and recall into a single metric, providing a balanced assessment of a classifier's performance. It ranges between 0 and 1, with higher values indicating better performance.
<br><br>


In a medical context, choosing the F1 score as the evaluation metric is often motivated by the need to balance the trade-off between precision and recall.
Both false positives and false negatives can have serious implications in healthcare, and the F1 score provides a comprehensive view of a model's ability to handle both types of errors. 

## <font color= grey>Test Classifier Algorithmn<a class="anchor" id="2.0"></a>

There are multiple text classifier algorithmn used for NLP. However, the most common one includes:<br><br>
1. **Naive Bayes**<br>

Based on Bayes' theorem, Naive Bayes assumes that features are conditionally independent given the class label. Despite its "naive" assumption, Naive Bayes often performs well for text classification tasks. It's computationally efficient and works particularly well with high-dimensional data like text.
<br><br>
2. **Support Vector Machine (SVM)**<br>

SVM is a powerful algorithm for both classification and regression tasks. In text classification, it works by finding a hyperplane that best separates data points belonging to different classes. SVMs are effective in high-dimensional spaces and can handle non-linear decision boundaries through the use of kernel functions.
<br><br>
3. **Logistic Regression**<br>

Despite its name, logistic regression is used for binary classification. It models the probability of a sample belonging to a particular class using the logistic function. Logistic regression is interpretable and works well when the relationship between features and the log-odds of the response variable is approximately linear.
<br><br>
4. **Random Forest**<br>

Random Forest is an ensemble method that constructs multiple decision trees during training and outputs the class that is the mode of the classes (classification) or mean prediction (regression) of the individual trees. It's robust, handles overfitting well, and can capture complex relationships in the data.
<br><br>
5. **Decision Tree**<br>

Decision trees recursively split the dataset based on features to create a tree-like structure. Each internal node represents a decision based on a feature, and each leaf node represents the class label. Decision trees are easy to understand, but they can be prone to overfitting.
<br><br>
6. **LightGBM**<br>

LightGBM is a gradient boosting framework that uses tree-based learning algorithms. It's designed for distributed and efficient training, making it particularly suitable for large datasets. LightGBM builds trees in a depth-wise fashion, prioritizing the growth of deeper trees to improve accuracy.
<br><br>

These algorithms are commonly used in natural language processing (NLP) tasks like text classification. The choice of algorithm depends on various factors, including the size of the dataset, the nature of the features, and the specific requirements of the task. For Hyper-parameter tuning, I have chosen to use optuna.Choosing Optuna for hyperparameter tuning over traditional methods like GridSearch and Random Search can offer several advantages:<br>

1. **Efficiency and Speed:** <br>Optuna employs an intelligent search strategy, exploring the hyperparameter space more efficiently than traditional methods. It dynamically allocates more trials to promising hyperparameter configurations, which can lead to faster convergence and better results.<br><br>

2. **Dynamic Hyperparameter Ranges:**<br> Optuna allows you to define dynamic ranges for hyperparameters, adjusting the search space during optimization based on the progress of the trials. This adaptability can be crucial for finding optimal values in complex optimization landscapes.<br><br>

3. **Adaptive Search:**<br> Optuna adapts its search strategy based on the outcomes of past trials. It uses information from completed trials to guide the search towards promising regions of the hyperparameter space, making it more likely to find optimal configurations.

```python
def plot_learning_curve(estimator, X_train, y_train, text_representation):
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X_train, y_train, cv=5, scoring='accuracy', n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10)
    )

    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Training Score')
    plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Validation Score')
    plt.title(f'Learning Curve ({text_representation})')
    plt.xlabel('Training Examples')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```
